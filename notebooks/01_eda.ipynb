```python
from src.tfidf_index import TFIDFIndex
import pandas as pd

# Load processed dataset
df = pd.read_csv("../data/processed/preprocessed_5000.csv")

docs = df["search_text"].tolist()

# Create TF-IDF index
tfidf = TFIDFIndex(docs)

# Sample query
query = "tomato onion rice"

indices, scores = tfidf.search(query, top_k=5)

print("Top results:\n")
for i, s in zip(indices, scores):
    print(f"{df.loc[i, 'name']}   (score: {s:.4f})")
```

```python
from src.bm25_index import BM25Index
import pandas as pd
from src.preprocessing import basic_clean, tokenize, remove_stopwords, lemmatize

# Load processed dataset
df = pd.read_csv("../data/processed/preprocessed_5000.csv")

# Use ingredients_tokens + steps_tokens for BM25
docs = df["ingredients_tokens"].apply(eval).tolist()  # convert string lists to actual lists

bm25 = BM25Index(docs)

# Preprocess query properly
query = "tomato onion ric"
clean = basic_clean(query)
tokens = tokenize(clean)
tokens = remove_stopwords(tokens)
query_tokens = lemmatize(tokens)

indices, scores = bm25.search(query_tokens, top_k=5)

for i, s in zip(indices, scores):
    print(df.loc[i, "name"], " -- score:", s)
```

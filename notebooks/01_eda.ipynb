```python
import pandas as pd

# Load the raw recipes dataset
recipes_path = "../data/raw/RAW_recipes.csv"   # adjust path if needed
df = pd.read_csv(recipes_path)

# Basic info
print("Number of recipes:", df.shape[0])
print("Number of columns:", df.shape[1])
print("\nColumns:", df.columns.tolist())

# Display first 5 rows
print("\nSample recipes:")
display(df.head())

# Show a complete sample recipe
sample = df.sample(1).iloc[0]

print("\n=== SAMPLE RECIPE ===")
print("Title:", sample['name'])
print("Minutes:", sample['minutes'])
print("Tags:", sample['tags'])
print("\nIngredients:", sample['ingredients'])
print("\nSteps:", sample['steps'])
```

```python
# Create a 5k sample dataset
sample_df = df.sample(5000, random_state=42)

# Save it for development
sample_path = "../data/sample/sample_5000.csv"
sample_df.to_csv(sample_path, index=False)

print("Sample dataset created at:", sample_path)
print("Shape:", sample_df.shape)
```
